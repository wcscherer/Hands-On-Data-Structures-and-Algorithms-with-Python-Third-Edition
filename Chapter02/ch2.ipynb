# Chapter 2 Questions: Algorithmic Complexity

Find the Big O time complexity of the following functions

```python
n = [1, 100, 1000, 100000, 1000000, 10000000]
```

```python
def timer(func):
    """
    This function is a decorator that times the execution of a function and returns the time it took to execute
    as well as the result of the function.
    """
    import time
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        return result, end - start
    return wrapper
```

```python
def plot_times(n, function):
    """
    This function plots the time it takes to run timer function with different values of n.
    """
    import matplotlib.pyplot as plt
    times = []
    for i in n:
        result, time = timer(function)(i)
        times.append(time)
    plt.plot(n, times)
    plt.xlabel("n")
    plt.ylabel("time")
    plt.title("Time complexity of function")
    plt.show()

```

## Problem 1)

```python
def problem1(n: int) -> int:
    i=1
    while(i<n): 
        i*=2 
        print("data") # the while loop is O(logn) since it doubles the value of i each iteration
```

Given that the loop iterator `i` is doubled each iteration, the loop will traverse `n` in log(n) time

Run the timer results and check against Big O

```python
plot_times(n, problem1)
```

## Problem 2)

```python
def problem2(n: int) -> int:
    i =n
    while(i>0):
        print('complexity')
        i /= 2
    return i # the while loop is O(logn) since it halves the value of i each iteration
```

Given that the iterator `i` starts at `n` and decrements by 1/2, the function will also execute in O(logn) time

```python
plot_times(n, problem2)
```

## Problem 3)

```python
def problem3(n: int) -> int:
    for i in range(1, n): # Outer loop is O(n)
        j = i
        while (j<n): #inner loop is O(logn)
            j *= 2
```

Given that the outer loop is O(n) and the inner loop is O(logn) the function is O(nlogn)

```python
plot_times(n, problem3)
```
